```python
# import general modules
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset # you don't need it until the 5th cell.

# plot inside the notebook
%matplotlib inline

# import the modules in this package
import gamap as ga

# block warnings to make this notebook cleaner.
import warnings
warnings.filterwarnings('ignore')
```

```python
# the simplest one-line tvmap example

# generate a fake 2D data
test_arr = np.outer(np.sin(np.linspace(0,1,30)*np.pi),
                    np.sin(np.linspace(0,1,60)*np.pi))
# remember that python's dimension order is (time,lev,lat,lon), with the fastest changing dimension on the right
Nlat,Nlon = np.shape(test_arr)
print(Nlat,Nlon)

ga.tvmap(test_arr) # one-line tvmap for quick check.
```

```python
# tvmap assumes a global grid by default.
# now pretend that test_arr is regional data over china
lon = np.linspace(60,150,Nlon,endpoint=False) # 60E~150E
lat = np.linspace(0,60,Nlat,endpoint=False) # 0~60N

# specify the lat and lon arrays, and some other parameters as well
ga.tvmap(test_arr,lon=lon,lat=lat,title='regional plot',unit='ppb')

# see all the parameters by help(ga.tvmap)
```

```python
# Now, go beyond the one-line tvmap. 
# To make subplots, you need to specify which 'axis' (i.e. sub-panel) to plot on.

# make 4 sub plots. You can control the figure size here.
f, axarr = plt.subplots(2, 2,figsize=(8,6)) # f is the entire figure and axarr contains all subplots

# plot on the first panel by specifying the axis.
ga.tvmap(test_arr,axis=axarr[0,0],title='subplot1',unit='') 

# plot on the second panel with the tvplot method. It doesn't assume lat-lon data.
ga.tvplot(test_arr,axis=axarr[0,1],title='subplot2',unit='') 

# specifying x, y explicitly for tvplot, otherwise it just uses array indices. 
# Also change a color scheme (for diff-plot). Otherwise, gamap's WhGrYlRd scheme is used.
x = np.linspace(0,100,Nlon) 
y = np.linspace(0,200,Nlat) 
ga.tvplot(test_arr,axis=axarr[1,0],x=x,y=y,cmap=plt.cm.RdBu_r,vmax=1,vmin=-1,
          xlabel='dimension1',ylabel='dimension2',title='subplot3',unit='') 

# make a totally different plot on the last panel
axarr[1,1].plot(np.arange(10),np.arange(10))
axarr[1,1].set_title('subplot4')

f.show() # Finally, show the figure interatively
```

```python
# now, plot GEOS-Chem diagnostics

# read data using the netCDF4 package
filename='sample_data/GEOSCHEM_Diagnostics_Hrly.201307020000.nc'
fh = Dataset(filename, "r", format="NETCDF4")

# fh.variables is a dictionary type. The variable names are the keys of the dictionary.
print(fh.variables.keys())
```

```python
# The data can be extracted directly by the key
print(fh['SPC_O3'].units)
data = fh['SPC_O3'][:]
print('Ntime,Nlev,Nlat,Nlon =',np.shape(data))
```

- Although the size of the 'time' dimension is just 1, it is still kept. So programming can be more consistent than with IDL. 

```python
# plot the surface ozone
ga.tvmap(data[0,0,:,:]*1e9,title='surface ozone',unit='ppb')
```

```python
# plot the tropospheric ozone profile
lat=fh['lat'][:]
profile = np.mean(data[0,0:30,:,:],axis=2)
ga.tvplot(profile*1e9,title='zonal mean ozone',unit='ppb',
          x=lat,xlabel='lat',ylabel='level')
```

```python
# always remember to close the NC file.
fh.close()
```
